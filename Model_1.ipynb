Barplots showing the frequency of each category separated by label

```python
#barplots showing the frequency of each category separated by label
plt.figure(figsize=[15,20])

fft=["department", "region","education", "gender","recruitment_channel","no_of_trainings", "age","previous_year_rating", "length_of_service", "awards_won?", "avg_training_score", "is_promoted"]                   
n=1
for f in fft:
    plt.subplot(12,1,n)
    sns.countplot(x=f, hue='is_promoted', edgecolor="black", alpha=0.7, data=train)
    sns.despine()
    plt.title("Countplot of {}  by is_promoted".format(f))
    n=n+1
plt.tight_layout()
plt.show()

```

Target
<li/>0  is not promoted
<li/>1  is promoted
<br/>As you can see, here we have imbalanced data, the number of 1 ( is promoted) < 0 (is not promoted)

```python
grouped = train.groupby(train.is_promoted)

gets_promoted = grouped.get_group(1)
```
